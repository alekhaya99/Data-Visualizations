```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
stack_data=pd.read_csv("D:/Data Science/365Data Science/Data VIs/Stacked/stacked_area_chart_data.csv")
```

```python
stack_data
```

```python
sns.set_palette('colorblind')
```

```python
colors=["#011638","#7e2987","ef2026"]
labels=["Diesel","Petrol","Gas"]
plt.figure(figsize=(12,6))
sns.set_style("white")
plt.stackplot(stack_data["Year"],stack_data["Diesel"],stack_data["Petrol"],stack_data["Gas"],
             colors=colors,edgecolor='none')
plt.xticks(stack_data["Year"],rotation=90)
plt.legend(labels=labels,loc="upper left")
plt.ylabel("Number of cars",fontsize=13)
plt.title("Popularity of Engine Fuel Types (1982-2016)",fontsize=25,fontweight="bold")
sns.despine()
plt.show()
```
