```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set()
```

```python
reg=pd.read_csv("D:/Data Science/365Data Science/Data VIs/Regression/scatter_plot_ii.csv")
```

```python
reg
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.regplot(x='Budget',
           y="Sales",
           data=reg,
           scatter_kws={'color':'blue'},
           line_kws={'color':'red'})
plt.title('Effect of Ad Expenditure on Sales (Regression)',fontsize=18,fontweight="bold")
plt.xlabel("Sales in (000's Units)",fontsize=13,fontweight="bold")
plt.ylabel("Price in (000's $)",fontsize=13,fontweight="bold")
plt.show()
```

```python

sns.lmplot(x='Budget',
           y="Sales",
           data=reg,
           scatter_kws={'color':'blue'},
           line_kws={'color':'red'},
          height=10)
plt.title('Effect of Ad Expenditure on Sales (Regression)',fontsize=18,fontweight="bold")
plt.xlabel("Sales in (000's Units)",fontsize=13,fontweight="bold")
plt.ylabel("Price in (000's $)",fontsize=13,fontweight="bold")
plt.show()
```
