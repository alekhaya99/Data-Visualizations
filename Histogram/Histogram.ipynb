```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
histogram_data=pd.read_csv("D:/Data Science/365Data Science/Data VIs/Histo/histogram_data.csv")
```

```python
histogram_data
```

```python
sns.set_style("white")
plt.figure(figsize=(10,10))
plt.hist(histogram_data["Price"],
        bins=8,
        color="blue")
plt.title("Distribution of Real Estate Prices",fontsize=18,fontweight="bold")
plt.xlabel("Price in (000 $)")
plt.ylabel("Number of Properties")
sns.despine()#Remove top and right borders
plt.show()
```
