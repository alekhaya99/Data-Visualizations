```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars=pd.read_csv("D:/Data Science/365Data Science/Data VIs/Bar Chart/bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize=(9,6))
plt.bar(x=df_used_cars["Brand"],height=df_used_cars["Cars Listings"],color='rgbwymc')
plt.xticks(rotation=45,fontsize=13)
plt.yticks(fontsize=13)
plt.title("Car Listing by Brand",fontsize=16,fontweight="bold")
plt.ylabel("Number of Cars Sold",fontsize=13)
plt.show()
plt.savefig("usedcarsbar.png")
```

```python
plt.figure(figsize=(15,15))
plt.barh(width=df_used_cars["Cars Listings"],y=df_used_cars["Brand"],color='blue')
plt.yticks(rotation=45,fontsize=13)
plt.xticks(fontsize=13)
plt.title("Car Listing by Brand",fontsize=16,fontweight="bold")
plt.xlabel("Number of Cars Sold",fontsize=13)
plt.show()
plt.savefig("usedcarsbar_hor.png")
```
